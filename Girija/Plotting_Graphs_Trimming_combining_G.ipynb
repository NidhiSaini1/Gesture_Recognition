```python
#FINAL CODE FOR PLOTTING MULTIPLE CSVs TOGETHER
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import io
from google.colab import files

# Upload multiple CSV files
uploaded_files = files.upload()

# Initialize an empty list to store DataFrames
dataframes = []

# Iterate through each uploaded file
for filename in uploaded_files.keys():
    # Read the CSV file into a DataFrame
    data = pd.read_csv(io.StringIO(uploaded_files[filename].decode('utf-8')))

    # Append the DataFrame to the list
    dataframes.append(data)

# Concatenate the DataFrames into one DataFrame
combined_data = pd.concat(dataframes, ignore_index=True)

# Plotting the graph
plt.figure(figsize=(15, 6))

# Plot each column against time
for column in combined_data.columns[1:]:
    plt.plot(combined_data['Time (s)'], combined_data[column], label=column)

# Adding Labels
plt.xlabel('Time (s)')
plt.ylabel('Acceleration')
plt.title('Line Graph of Accelerations vs Time')

# For Legend
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```
